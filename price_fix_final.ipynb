```python
import math

def calculate_probability(num_attempts, num_successes):
    p = 0.007815735
    q = 1 - p
    return p ** num_successes * (q ** (num_attempts - num_successes)) * math.comb(num_attempts, num_successes)

def calculate_and_print_results(num_attempts, num_successes):
    total_sum = 0
    for i in range(1, num_attempts + 1):
        result = calculate_probability(i, num_successes)
        total_sum += result
    print("Total sum of the results:", total_sum)

# Call the function for different scenarios
calculate_and_print_results(433, 1)  # 2 attempts, 1 success
calculate_and_print_results(648, 2)  # 3 attempts, 2 successes
calculate_and_print_results(865, 3)  # 4 attempts, 3 successes

# p 성공확률 , calculate_and_print_results(a, i) , a  = 판매가격*(1-수수료) / 1회당 강화비용
```

```python
# Given values
initial_amount = 1000
probability_of_winning = 0.10
cost_per_game = 50

# Calculate expected value
amount_won = initial_amount - cost_per_game
probability_of_losing = 1 - probability_of_winning
amount_lost = cost_per_game
expected_value = (probability_of_winning * amount_won) + (probability_of_losing * amount_lost)

# Calculate the percentage of loss
percentage_of_loss = (amount_lost / expected_value) * 100

print("Expected Value:", expected_value)
print("Percentage of Loss:", percentage_of_loss)
```
